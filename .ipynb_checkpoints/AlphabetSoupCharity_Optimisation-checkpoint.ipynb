```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import tensorflow as tf
import sklearn as skl
import keras_tuner as kt
from pprint import pprint
from tensorflow.keras.models import load_model, save_model
```

```python
#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df.info()
```

```python
# Copy the original df
application_copy_df =  application_df.copy()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_copy_df = application_copy_df.drop(columns=['EIN', 'NAME'])
application_copy_df
```

```python
# Determine the number of unique values in each column.
application_copy_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_copy_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cut_off = 200
application_types_to_replace = application_copy_df['APPLICATION_TYPE'].value_counts()\
                                [application_copy_df['APPLICATION_TYPE'].value_counts() < cut_off]\
                                .index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_copy_df['APPLICATION_TYPE'] = application_copy_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_copy_df['APPLICATION_TYPE'].value_counts()
```

```python
# Check the items in Other
application_copy_df['APPLICATION_TYPE'].value_counts()\
                                [application_copy_df['APPLICATION_TYPE'].value_counts() < cut_off]
```

```python
# (Alternative solution) Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cut_off = 200
application_types_to_replace = ['T9','T13','T12','T2','T25','T14','T29','T15','T17']

# Replace in dataframe
for app in application_types_to_replace:
    application_copy_df['APPLICATION_TYPE'] = application_copy_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_copy_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_copy_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
cut_off = 800
classifications_to_replace = application_copy_df['CLASSIFICATION'].value_counts()\
                                [application_copy_df['CLASSIFICATION'].value_counts() < cut_off]\
                                .index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_copy_df['CLASSIFICATION'] = application_copy_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_copy_df['CLASSIFICATION'].value_counts()
```

```python
# Return all the object columns
application_copy_df.dtypes[application_copy_df.dtypes == "object"]
```

```python
# Generate categorical variable list
application_cat = application_copy_df.dtypes[application_copy_df.dtypes == "object"].index.tolist()
application_cat
```

```python
# Check unique values in the application_cat list
application_copy_df[application_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_copy_df[application_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(application_cat)
encode_df.head()
```

```python
application_copy_df
```

```python
# Merge one-hot encoded features and drop the originals
application_encoded_df = application_copy_df.merge(encode_df,left_index=True, right_index=True)
application_encoded_df = application_encoded_df.drop(application_cat, axis=1)
application_encoded_df
```

```python
# Check all columns of the encoded dataframe
application_encoded_df.columns
```

```python
# Split our preprocessed data into our features and target arrays
y = application_encoded_df['IS_SUCCESSFUL'].values
X = application_encoded_df.drop('IS_SUCCESSFUL', axis = 1).values
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model
## Test model 1 - reducing the number of neurons for each layer to 5

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0]) # number of columns
hidden_nodes_layer1 =  9

nn1 = tf.keras.models.Sequential()

# First hidden layer
nn1.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Output layer
nn1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn1.summary()
```

```python
# Compile the model
nn1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn1.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn1.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

## Test model 2 - adding an extra layer

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0]) # number of columns
hidden_nodes_layer1 =  5
hidden_nodes_layer2 = 5

nn2 = tf.keras.models.Sequential()

# First hidden layer
nn2.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn2.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Third hidden layer
nn2.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))


# Output layer
nn2.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn2.summary()
```

## Model optimisation

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp): #hyperparameter search object, hp refers to Hyperband in the cell below
    # when you call the create_model function within kt.Hyperband(), it will automatically pass in Hyperband in the function

    nn_model = tf.keras.models.Sequential()# finding the optimal hyperparameters for a Sequential model

    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','sigmoid']) #allow Hyperband to select between different activation
    #functions during tuning process

    # Allow kerastuner to decide number of neurons in first layer
    #'first_units' is the name of the hyperparameter
    #name is used internally by the tuner to keep track of which hyperparameter it is adjusting and to log or
    #report the values it is testing
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=10,
        step=2), activation=activation, input_dim=len(X_train[0]))) #increment in steps of 2

    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 6)):
      units_name = 'units_' + str(i)
      print(f"Adding layer with {units_name}") # for debugging purposes
      nn_model.add(tf.keras.layers.Dense(units=hp.Int(units_name,
          min_value=1,
          max_value=10,
          step=2),
          activation=activation))



    # Define the output layer
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

    # Compile the model
    # can also use the tuner for loss function and optimiser as well
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])

    return nn_model
```

```python
tuner = kt.Hyperband( #initialise a hyperband tuner
    create_model, #use the create_model function, this tells the Hyperband what is the architecture of your model
    objective="val_accuracy", # when searching for the optimal hyperparameter, assess it against the validation accuracy
    # what's the accuracy of the current model with the hyperparameters with the test set?
    max_epochs=20, #20 epochs for the hyperparameter tuning process => for each combination (e.g., relu with 10 neurons, run
    # 20 times)
    hyperband_iterations=2)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

## Return the best hyperparameter model
