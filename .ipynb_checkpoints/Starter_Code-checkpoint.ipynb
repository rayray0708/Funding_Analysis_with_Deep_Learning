## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
application_df.info()
```

```python
# Copy the original df 
application_copy_df =  application_df.copy()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_copy_df = application_copy_df.drop(columns=['EIN', 'NAME'])
application_copy_df
```

```python
# Determine the number of unique values in each column.
application_copy_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_copy_df['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cut_off = 200 
application_types_to_replace = application_copy_df['APPLICATION_TYPE'].value_counts()\
                                [application_copy_df['APPLICATION_TYPE'].value_counts() < cut_off]\
                                .index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_copy_df['APPLICATION_TYPE'] = application_copy_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_copy_df['APPLICATION_TYPE'].value_counts()
```

```python
# Check the items in Other
application_copy_df['APPLICATION_TYPE'].value_counts()\
                                [application_copy_df['APPLICATION_TYPE'].value_counts() < cut_off]
```

```python
# (Alternative solution) Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cut_off = 200
application_types_to_replace = ['T9','T13','T12','T2','T25','T14','T29','T15','T17']

# Replace in dataframe
for app in application_types_to_replace:
    application_copy_df['APPLICATION_TYPE'] = application_copy_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_copy_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_copy_df['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
cut_off = 800
classifications_to_replace = application_copy_df['CLASSIFICATION'].value_counts()\
                                [application_copy_df['CLASSIFICATION'].value_counts() < cut_off]\
                                .index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_copy_df['CLASSIFICATION'] = application_copy_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_copy_df['CLASSIFICATION'].value_counts()
```

```python
# Return all the object columns
application_copy_df.dtypes[application_copy_df.dtypes == "object"]
```

```python
# Generate categorical variable list
application_cat = application_copy_df.dtypes[application_copy_df.dtypes == "object"].index.tolist()
application_cat
```

```python
# Check unique values in the application_cat list
application_copy_df[application_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False) 

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(application_copy_df[application_cat])) 

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(application_cat) 
encode_df.head()
```

```python
application_copy_df
```

```python
# Merge one-hot encoded features and drop the originals
application_encoded_df = application_copy_df.merge(encode_df,left_index=True, right_index=True)
application_encoded_df = application_encoded_df.drop(application_cat, axis=1)
application_encoded_df
```

```python
# Check all columns of the encoded dataframe
application_encoded_df.columns 
```

```python
# Split our preprocessed data into our features and target arrays
y = application_encoded_df['IS_SUCCESSFUL'].values
X = application_encoded_df.drop('IS_SUCCESSFUL', axis = 1).values
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0]) # number of columns
hidden_nodes_layer1 =  8
hidden_nodes_layer2 = 8

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
